```python
import random

# Функция для определения выигрыша по выпавшему числу
def calculate_win(number):
    if number == 777:
        return 200
    elif number == 999:
        return 100
    elif number == 555:
        return 50
    elif number == 333:
        return 15
    elif number == 111:
        return 10
    elif number % 100 == 77 or number % 10 == 7:
        return 5
    elif number % 10 == 7:
        return 3
    elif number % 100 == 0 or number % 10 == 0:
        return 1
    else:
        return 0

# Функция для симуляции одной игры
def play_game():
    # Выпадение случайного числа от 0 до 999
    number = random.randint(0, 999)
    # Рассчитываем выигрыш
    win = calculate_win(number)
    # Игрок платит 1 рубль, поэтому выигрыш или проигрыш рассчитывается как win - 1
    return win - 1

# Симулируем большое количество игр
num_games = 1000000
total_win = sum(play_game() for _ in range(num_games))

# Рассчитываем средний выигрыш или проигрыш за одну игру
average_win = total_win / num_games

print(f"Средний выигрыш или проигрыш за одну игру: {average_win:.4f} руб.")

# Определяем, выгодна ли игра для игрока
if average_win > 0:
    print("Игра выгодна для игрока.")
elif average_win < 0:
    print("Игра не выгодна для игрока.")
else:
    print("Игра нейтральна для игрока.")

### Теоретический расчет среднего выигрыша

# Теоретический расчет среднего выигрыша
theoretical_win = (
    (1/1000) * 200 +
    (1/1000) * 100 +
    (1/1000) * 50 +
    (1/1000) * 15 +
    (1/1000) * 10 +
    (10/1000) * 5 +  # Для *77
    (10/1000) * 3 +  # Для *7
    (10/1000) * 2 +  # Для *00
    (10/1000) * 1   # Для *0
) - 1

print(f"Теоретический средний выигрыш или проигрыш за одну игру: {theoretical_win:.4f} руб.")

# Определяем, выгодна ли игра для игрока по теоретическому расчету
if theoretical_win > 0:
    print("Игра выгодна для игрока по теоретическому расчету.")
elif theoretical_win < 0:
    print("Игра не выгодна для игрока по теоретическому расчету.")
else:
    print("Игра нейтральна для игрока по теоретическому расчету.")

    # Сравнение результатов по знаку и числовой разнице
if (average_win > 0 and theoretical_win > 0):
    print("Оба результата имеют положительные значения.")
elif (average_win < 0 and theoretical_win < 0):
    print("Оба результата имеют отрицательные значения.")
else:
    print("Результаты имеют разные значения.")

# Числовая разница между результатами
difference = abs(average_win - theoretical_win)
print(f"Числовая разница между результатами: {difference:.4f} руб.")

# Определение, какой результат больше или меньше
if average_win > theoretical_win:
    print("Результат моделирования больше, чем теоретический результат.")
elif average_win < theoretical_win:
    print("Результат моделирования меньше, чем теоретический результат.")
else:
    print("Результаты моделирования и теоретические результаты равны.")

# Рекомендация, на какой результат ориентироваться
if abs(average_win) > abs(theoretical_win):
    print("Стоит ориентироваться на результат моделирования.")
elif abs(average_win) < abs(theoretical_win):
    print("Стоит ориентироваться на теоретический результат.")
else:
    print("Оба результата имеют одинаковую абсолютную величину, можно ориентироваться на любой из них.")
```
