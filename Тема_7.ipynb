```python
# Импортируем необходимые библиотеки
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

# Загрузка данных из CSV-файла
csv_file_path = '/content/web_clients_correct_доп_материал_к_тема_6__7__итоговое_задание__1_ (1).csv'
df = pd.read_csv(csv_file_path)

# Проверяем первые 15 строк данных
df_subset = df.head(15)

# Создание диаграммы разброса с помощью Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='bill')
plt.title('Зависимость расходов от возраста клиента')
plt.xlabel('Возраст')
plt.ylabel('Расходы')
plt.show()

# Создание диаграммы разброса с помощью Plotly
fig = px.scatter(df, x='age', y='bill')
fig.update_layout(title='Зависимость расходов от возраста клиента',
                  xaxis_title='Возраст',
                  yaxis_title='Расходы')
fig.show()
```
